```python
import os
import json
import cv2
import random
import yaml
from tqdm import tqdm
import matplotlib.pyplot as plt
import random
import torch
import numpy as np

def seed_everything(seed=42):
    # Python
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

    # NumPy
    np.random.seed(seed)

    # PyTorch CPU
    torch.manual_seed(seed)

    # PyTorch GPU
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    # Đảm bảo deterministic
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# Gọi hàm ngay khi khởi tạo notebook
seed_everything(42)
```

## 3. Training mô hình YOLOv11

Import thư viên để train mô hình YOLOv11

```python
from ultralytics import YOLO
```

Tải mô hình YOLOv11 gốc (yolo11s.pt, yolo11n.pt, ...)

```python
model = YOLO('yolo11s.pt') # Hoặc yolo11n.pt, ...
```

Thực hiện train mô hình với các tham số tùy chỉnh và augmentation

Nên kiểm tra file dataset.yml xem đã đúng đường dẫn các file train, val, nc, names chưa. Do có thể train trên môi trường có đường dẫn khác nhau nên cần kiểm tra kỹ (Colab, Kaggle, local, ...)

```python
# Huấn luyện mô hình với các tham số tùy chỉnh và augmentation
# Thay đổi đường dẫn data nếu cần thiết
results = model.train(
        data='/content/drive/MyDrive/3_HK1-25_26/DeepLearning/ZaloAIC2025/Track1/yolo_dataset/dataset.yaml',
        epochs=100,
        seed=42,
        imgsz=640,
        batch=64,
        lr0=0.001,

        # Add augmentation
        mixup=0.1,
        degrees=15.0,
        shear=5.0,

        name='drone_training'
    )
```

Sau khi train xong thì tải file best.pt để thực hiện bước inference
